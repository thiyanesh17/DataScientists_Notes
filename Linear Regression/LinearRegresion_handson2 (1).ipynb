```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_enhanced_pie(series, title="Distribution", figsize=(4, 4), colors=plt.cm.Pastel1.colors, save_path=None):
    def autopct_format(pct, allvals):
        absolute = int(round(pct / 100. * sum(allvals)))
        return f"{pct:.1f}%\n({absolute})"

    explode = [0.05] * len(series)

    plt.figure(figsize=figsize)
    wedges, texts, autotexts = plt.pie(
        series,
        labels=series.index,
        autopct=lambda pct: autopct_format(pct, series),
        colors=colors,
        explode=explode,
        startangle=140,
        shadow=True,
        textprops={'fontsize': 11, 'color': 'black'}
    )

    for text in texts:
        text.set_fontsize(12)
        text.set_color('darkblue')

    plt.title(title, fontsize=14, fontweight='bold', color='darkgreen')
    plt.axis('equal')
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=100, bbox_inches='tight')

    plt.show()

```
