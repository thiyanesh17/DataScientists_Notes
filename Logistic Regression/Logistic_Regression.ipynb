```python
import pandas as pd
titanic = pd.read_csv('/content/Simple_Train.csv')
titanic.head()
```

```python
Survived: Survived (1) or died (0); this is the target variable
Pclass: Passenger's class (1st, 2nd or 3rd class)
Name: Passenger's name
Sex: Passenger's sex
Age: Passenger's age
SibSp: Number of siblings/spouses aboard
Parch: Number of parents/children aboard
Ticket: Ticket number
Fare: Fare
Cabin: Cabin
Embarked: Port of embarkation
```

```python
ports = pd.get_dummies(titanic.Embarked, prefix='Embarked')
ports.head()
```

```python
titanic.isna().sum()
```

```python
#titanic['Embarked'].dropna(inplace=True)

titanic.dropna(subset=['Embarked'],inplace=True)
```

```python
titanic['Embarked'].isna().sum()
```

```python
titanic = titanic.join(ports)
titanic
```

```python
titanic.drop(columns=['Embarked'],inplace=True)
```

```python
titanic.head()
```

```python
titanic.Sex = titanic.Sex.map({'male':0,'female':1})
```

```python
y = titanic.Survived.copy()
X= titanic.drop(columns=['Survived'])

#Cabin, Ticket, Name,Passenger_ID

X.drop(columns =['Cabin','Name','PassengerId','Ticket'],inplace=True)
X.info()
```

```python
X.Age.fillna(X.Age.mean(),inplace=True)
```

```python
X.info()
```

```python
X.isna().sum()
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=100)
model = LogisticRegression()

model.fit(X_train,y_train)
```

```python
print("train",model.score(X_train,y_train))

print("test",model.score(X_test,y_test))
```
